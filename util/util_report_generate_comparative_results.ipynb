```python
from tqdm import tqdm
from glob import glob
from pathlib import Path

filenames_baseline = glob('./valid_baseline/*.jpg')
filenames_baseline_moretraining = glob('./valid_baseline_moretraining/*.jpg')
filenames_yolo = glob('./valid_yolov5/*.jpg')

```

```python
filenames_yolo
```

```python
html = '''
<!DOCTYPE html>
<html lang="pt-BR">
<head><title>Comparing validation images</title></head>
<body>
    <table>
        <tbody>
            <tr>
                <td>Baseline</td>
                <td>Baseline with more training</td>
                <td>Yolov5 best model</td>
            <tr>
            <tr>
                <td>Subaparts PPN (SSD). 100 epochs. Network input 480x480px.</<td>
                <td>Subaparts PPN (SSD). 1430 epochs. Dataset v3. Network input 480x480px.</td>
                <td>YOLOv5 small. Dataset v3. 300 epochs. Network input 640x640px.</td>
            </tr>
            

'''

def addImageRow(basline, baseline_moretraining, yolov5):
    ret = f'''
    <tr>
        <td><img src="{basline}" alt="erro" loading="lazy"  width="400" height="400"></td>
        <td><img src="{baseline_moretraining}" alt="erro" loading="lazy"  width="400" height="400"></td>
        <td><img src="{yolov5}" alt="erro" loading="lazy" width="400" height="400"></td>
    <tr>
    '''
    return ret


for filename in filenames_baseline:
    stem = Path(filename).stem
    yolo_f = ''
    for yolo_file in filenames_yolo:
        if stem in yolo_file:
            yolo_f = Path(yolo_file).stem
            break
    print(f'{stem}.jpg -> {yolo_file}.jpg')
    prefix  = '.'
    baseline_path = f'{prefix}/valid_baseline/{stem}.jpg'
    baseline_moretraing_path = f'{prefix}/valid_baseline_moretraining/{stem}.jpg'
    yolov5_path = f'{prefix}/valid_yolov5/{yolo_f}.jpg'
    html += addImageRow(baseline_path, baseline_moretraing_path, yolov5_path)


html += "</body></html)"

file = open('index.html', 'w',encoding="utf-8")
file.write(html)
```
